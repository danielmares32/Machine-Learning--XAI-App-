```python
import pandas as pd
import shap
from pycaret.classification import *
```

# Cargar los datos:

Este dataset es sobre enfermedades del corazón, el objetivo es predecir si la persona tiene una enfermedad o no.

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

# Entrenando y evaluando el modelo (Establecer el experimento)

```python
cat_features = ['sex','cp','fbs','restecg','exang','slope','ca','thal']
```

```python
experiment = setup(df, target='target', categorical_features=cat_features)
```

Se selecciona el mejor modelo de Machine Learning, es el que tiene mejor Accuracy

```python
compare_models()
```

```python
best_model = create_model('rf')
```

# Probamos el Modelo

```python
predict_model(best_model, df.tail())
```

Podemos observar que la columna Label es la predicción en el experimento. Si es 0 significa que la persona no tiene la enfermedad y si es 1 significa que si la tiene. En la columna target es la realidad, si la persona tenga o no tenga.

# Guardamos el Modelo

```python
save_model(best_model, model_name='rf')
```

# Interpretamos mediante SHAP (XAI)

```python
model = load_model('model')
train_pipe = model[:-1].transform(df)
explainer = shap.TreeExplainer(model.named_steps["trained_model"])
shap_values = explainer.shap_values(train_pipe)
interpret_model(best_model,plot='reason')

exp = explainer.explain_instance(X_test.values[j], knn.predict, num_features=5)
exp.show_in_notebook(show_table=True)
#shap.force_plot(explainer.expected_value[0], shap_values[0], train_pipe)
```
